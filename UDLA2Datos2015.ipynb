# Dataset del año 2015

## Importar datos

```python
import pandas as pd
import numpy as np
```

```python
df2015 = pd.read_excel('./drive/MyDrive/DatosInvestigacion/FINAL_resumen cuenca ago 2017.xlsx',sheet_name="PROMEDIO_2015")
```

```python
df2015 = df2015.iloc[1:][:]
```

```python
df2015
```

## Datos nulos

Existe una fila de datos nulos.

```python
np.where(df2015.isna())
```

```python
df2015.iloc[3270:3275][:]
```

Completar valores perdidos o faltantes. Se usa el método front filling.

```python
df_completed = df2015.fillna(method='ffill')
```

```python
df_completed.iloc[3270:3275][:]
```

Descripción de los datos.

```python
df_completed.describe()
```

```python
df_completed.dtypes
```

**Fechas**

```python
df_completed['Date'].describe(datetime_is_numeric=True)
```

Lista con datos de fecha y hora unidos.

```python
import datetime
```

```python
dateTimeIndex = [datetime.datetime.combine(df_completed['Date'][i],df_completed['Time'][i]) for i in df_completed.index ]
df_completed['DateTime'] = dateTimeIndex
```

```python
df_completed['DateTime'].head()
```

## Visualización de la Radiación Solar Global Promedio ( SRGLOAVG1H )

La descripción muestra que el 50% de los datos están por debajo de 7 ¿unidades? W/m2 o MJ/m2.

Esto es porque la mitad de los datos están en la noche.

```python
df_completed['SRGLOAVG1H'].describe()
```

```python
import matplotlib.pyplot as plt
```

Se colocan las fechas como índices

```python
df_GSI = df_completed.set_index('DateTime')
```

```python
df_GSI.SRGLOAVG1H.plot(figsize=(20,5), title = "Global Solar Irradiation")
```

**Datos del mes de enero**

```python
enero2015 = df_GSI.SRGLOAVG1H['2015-01']
```

```python
enero2015.plot(figsize=(20,5), title = "Global Solar Irradiation in one month")
```

**Datos de un día en específico**

```python
enero10_2015 = df_GSI.SRGLOAVG1H['2015-01-10']
```

```python
enero10_2015.plot(figsize=(20,5), title = "Global Solar Irradiation in one day")
```

**Datos de una semana**

Puede verse el incremento de la radiación solar global durante el día y su disminución en la noche.

```python
semana_enero2015 = df_GSI.SRGLOAVG1H['2015-01-11':'2015-01-18']
```

```python
semana_enero2015.plot(figsize=(20,5), title = "Global Solar Irradiation in one day")
```
