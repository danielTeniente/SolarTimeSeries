# Resultados

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
import seaborn as sns
```

```python
df_predicciones = pd.read_csv('./predicciones_modelos.csv')
```

```python
df_predicciones.head()
```

```python
real = df_predicciones['Real Data']
AR = df_predicciones['AR']
LSTM_u = df_predicciones['LSTM univariado']
LSTM_m = df_predicciones['LSTM multivariado']
```

## AR

```python
sns.scatterplot(x=real,y=AR)
plt.plot(real, real, linewidth=2,color="r")
```

```python
print('RMSE:', np.sqrt(mean_squared_error(real,AR)))
print('MAE:', mean_absolute_error(real,AR))
print('R_2:', r2_score(real,AR))
```

## LSTM univariado

```python
sns.scatterplot(x=real,y=LSTM_u)
plt.plot(real, real, linewidth=2,color="r")
```

```python
print('RMSE',np.sqrt(mean_squared_error(real,LSTM_u)))
print('MAE',mean_absolute_error(real,LSTM_u))
print('R_2',r2_score(real,LSTM_u))
```
